```
%sql
WITH
CTE_TRIPNUM AS(
  SELECT
        DATE_FORMAT(pickup_datetime,'MMM y') AS year_month,
        COUNT(*) as trip_num
        FROM yellow_green_merged
        GROUP BY year_month
),

CTE_MOST_DAY_OF_MONTH AS(
  SELECT
        year_month,
        day_of_month
        FROM(
            SELECT
                  year_month,
                  trip_num,
                  ROW_NUMBER() OVER (PARTITION BY year_month ORDER BY trip_num DESC) as RNK,
                  day_of_month
                  FROM(
                      SELECT
                            DATE_FORMAT(pickup_datetime,'MMM y') as year_month,
                            COUNT(*) as trip_num,
                            DATE_FORMAT(pickup_datetime,'EEEE') as day_of_month
                            FROM yellow_green_merged
                            GROUP BY year_month, day_of_month
                            ORDER BY year_month
                  )
        )
        WHERE RNK = 1
),

CTE_MOST_HOUR_OF_MONTH AS(
  SELECT
        year_month,
        hour_of_month
        FROM(
            SELECT
                  year_month,
                  trip_num,
                  ROW_NUMBER() OVER (PARTITION BY year_month ORDER BY trip_num DESC) as RNK,
                  hour_of_month
                  FROM(
                      SELECT
                            DATE_FORMAT(pickup_datetime,'MMM y') as year_month,
                            COUNT(*) as trip_num,
                            DATE_FORMAT(pickup_datetime,'HH') as hour_of_month
                            FROM yellow_green_merged
                            GROUP BY year_month, hour_of_month
                            ORDER BY year_month
                  )
        )
        WHERE RNK = 1
),

CTE_AVG_CALC AS(
  SELECT
        DATE_FORMAT(pickup_datetime,'MMM y') AS year_month,
        ROUND((SUM(passenger_count)/count(*)),2) as avg_passener,
        ROUND((SUM(total_amount)/count(*)),2) as avg_paid_amount,
        ROUND((SUM(total_amount)/count(passenger_count)),2) as avg_payment_by_passener
        FROM yellow_green_merged
        GROUP BY year_month
        ORDER BY year_month
)


SELECT 
CTN.year_month,
CTN.trip_num,
CMDM.day_of_month,
CMHM.hour_of_month,
CAC.avg_passener,
CAC.avg_paid_amount,
CAC.avg_payment_by_passener
FROM
CTE_TRIPNUM AS CTN
JOIN CTE_MOST_DAY_OF_MONTH AS CMDM
ON CTN.year_month = CMDM.year_month
JOIN CTE_MOST_HOUR_OF_MONTH AS CMHM
ON CTN.year_month = CMHM.year_month
JOIN CTE_AVG_CALC AS CAC
ON CTN.year_month = CAC.year_month
ORDER BY year_month
```

```
%sql
SELECT
cab_type,
ROUND(AVG(datediff(SECOND, pickup_datetime, dropoff_datetime)/60),2) as avg_trip_duration,
ROUND(percentile_approx((datediff(SECOND, pickup_datetime, dropoff_datetime)/60),0.5),2) as median_trip_duration,
ROUND(MIN(datediff(SECOND, pickup_datetime, dropoff_datetime)/60),2) as minimum_trip_duration,
ROUND(MAX(datediff(SECOND, pickup_datetime, dropoff_datetime)/60),2) as maximum_trip_duration,
ROUND(AVG(trip_distance*1.61),2) as avg_distance_km,
ROUND(PERCENTILE_APPROX((trip_distance*1.60934),0.5),2) as median_distance_km,
ROUND(MIN(trip_distance*1.61),2) as minimum_distance_km,
ROUND(MAX(trip_distance*1.61),2) as maximum_distance_km,
ROUND(AVG((trip_distance*1.61)/(datediff(SECOND, pickup_datetime, dropoff_datetime)/3600)),2) as avg_speed,
ROUND(PERCENTILE_APPROX(((trip_distance*1.61)/(datediff(SECOND, pickup_datetime, dropoff_datetime)/3600)),0.5),2) AS median_speed,
ROUND(MIN((trip_distance*1.61)/(datediff(SECOND, pickup_datetime, dropoff_datetime)/3600)),2) as minimum_speed,
ROUND(MAX((trip_distance*1.61)/(datediff(SECOND, pickup_datetime, dropoff_datetime)/3600)),2) as maximum_speed
FROM yellow_green_merged
GROUP BY cab_type

--
--
-- 1 means yellow
-- 2 means green
```

```
%sql
-- Add a new column named bins
WITH CTE_tripdur AS(
SELECT
datediff(SECOND, pickup_datetime, dropoff_datetime)/3600 as trip_duration_hour,trip_distance,total_amount,
CASE
       WHEN datediff(SECOND, pickup_datetime, dropoff_datetime)/60<5 then 'under 5 minutes'
       WHEN datediff(SECOND, pickup_datetime, dropoff_datetime)/60>=5 and datediff(SECOND, pickup_datetime, dropoff_datetime)/60<10 then 'from 5 minutes to 10 minutes' 
       WHEN datediff(SECOND, pickup_datetime, dropoff_datetime)/60>=10 and datediff(SECOND, pickup_datetime, dropoff_datetime)/60<20 then 'from 10 minutes to 20 minutes' 
       WHEN datediff(SECOND, pickup_datetime, dropoff_datetime)/60>=20 and datediff(SECOND, pickup_datetime, dropoff_datetime)/60<30 then 'from 20 minutes to 30 minutes' 
       WHEN datediff(SECOND, pickup_datetime, dropoff_datetime)/60>=30 and datediff(SECOND, pickup_datetime, dropoff_datetime)/60<60 then 'from 30 minutes to 60 minutes' 
       WHEN datediff(SECOND, pickup_datetime, dropoff_datetime)/60>=60 then 'at least 60 minutes' 
       ELSE NULL
       END as bins
FROM yellow_green_merged
)
SELECT
bins,
ROUND(AVG((trip_distance*1.61)/trip_duration_hour),2) as AVG_SPEED,
ROUND(AVG((trip_distance*1.61)/total_amount),2) as avg_distance_per_dollar
FROM CTE_tripdur
GROUP BY bins
ORDER BY bins DESC
```

```
from pyspark.ml.feature import OneHotEncoder, StringIndexer, VectorAssembler


stages = []
```
